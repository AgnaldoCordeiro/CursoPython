```python
import pandas as pd
import pandera as pa
```

```python
valores_ausentes = ['###!', '**', '####', '****', 'NULL', '*****']
df = pd.read_csv("ocorrencia.csv", encoding="ISO-8859-1", engine="python", sep=';', parse_dates=['ocorrencia_dia'], dayfirst=True, na_values=valores_ausentes)
df.head(10)
```

```python
schema = pa.DataFrameSchema(
    columns={
         "codigo":pa.Column(pa.Int, required=False),#isso é para caso ocorra uma coluna vir outra hora nao ai colcoar o required false pois vem padrao true
        "codigo_ocorrencia":pa.Column(pa.Int),
        "codigo_ocorrencia2": pa.Column(pa.Int),
        "ocorrencia_classificacao": pa.Column(pa.String),
        "ocorrencia_cidade": pa.Column(pa.String),
        "ocorrencia_uf": pa.Column(pa.String, pa.Check.str_length(2,2), nullable=True), #joga um parametro para a minimo e maximo de tamanho
        "ocorrencia_aerodromo": pa.Column(pa.String, nullable=True),
        "ocorrencia_dia": pa.Column(pa.DateTime),
        "ocorrencia_hora": pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
        "total_recomendacoes": pa.Column(pa.Int)
        #str_matches é uma expressao regular de validação de hora, porem faz muito mais coisas tbm pesquisar
    }
)
```

```python
schema.validate(df)
```

```python
#ocorrencias cujo a cidade ocmencem com a Letra C
filtro = df.ocorrencia_cidade.str[0] == 'A' # para ver que letra termina só usar no indice -1
df.loc[filtro]
```

```python
#outro metodo de acesso dentro da string
#filtro = df.ocorrencia_cidade.str.contains('MA') #verifica se contem em qualquer parte do conteudo
filtro = df.ocorrencia_cidade.str.contains('MA | AL')# verificar se tem um ou o outro
df.loc[filtro]
```

```python
#filtro de data
#filtro = df.ocorrencia_dia.dt.year == 2015
filtro = (df.ocorrencia_dia.dt.year == 2015) & (df.ocorrencia_dia.dt.month == 12) & (df.ocorrencia_dia.dt.day >= 3) & (df.ocorrencia_dia.dt.day <=11)
df.loc[filtro]
```

```python
#criando novos dados ou seja juntando colunas por exemplo
df['ocorrencia_dia_hora'] = pd.to_datetime(df.ocorrencia_dia.astype(str) + ' ' + df.ocorrencia_hora)
#o to_datetime pega a informação ocnvertida em texto pelo astype(str) pois nao tem como juntar um tipo date com string e depois que juntou ele transforma em datetime
```
